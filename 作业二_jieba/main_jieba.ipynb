# 结巴分词实践

## 结巴分词的三种模式

### 精确模式→Shift+Enter

```python
import jieba  # 导入结巴库，所有分词代码都要先导入
# 精确模式：分词结果最精准，无冗余，默认模式
sentence1 = "我来到北京清华大学"
seg_exact = jieba.lcut(sentence1, cut_all=False)  # cut_all=False是精确模式
print("精确模式分词结果：", seg_exact)
```

```python
# 全模式：列出所有可能的分词结果，有冗余
seg_all = jieba.lcut(sentence1, cut_all=True)
print("全模式分词结果：", seg_all)
```

```python
# 搜索引擎模式：对长词再次切分，适合搜索引擎
sentence2 = "小明硕士毕业于中国科学院计算所，后在日本京都大学深造"
seg_search = jieba.lcut_for_search(sentence2)
print("搜索引擎模式分词结果：", seg_search)
```

```python
# 载入自定义词典
jieba.load_userdict("userdict.txt")
# 验证姓名分词
my_name = "ropz即将是自然语言处理方面的高手。"
seg_name = jieba.lcut(my_name, HMM=False)
print("姓名分词验证结果：", seg_name)
```
